#### PROV-TEMPLATE Spec available at https://openprovenance.org/prov-template-2014-06-07/

Prov Preamble

```python
from prov.serializers.provjson import ProvJSONSerializer
from prov.serializers.provxml import ProvXMLSerializer
from IPython.display import Image
import io
from prov.model import ProvDocument, ProvBundle, ProvException, first, Literal
from prov.dot import prov_to_dot


def provXMLtoPNG(doc,format):
     
    
    doc=ProvXMLSerializer.deserialize(ProvXMLSerializer,doc)
    dot = prov_to_dot(doc)
    
    if format=="png":
        dot.write_png('PROV.png')
        return 'PROV.png' 
    if format=="svg":
        dot.write_svg('PROV.svg')
        return 'PROV.svg'

def provJSONtoPNG(doc,format):
     
    
    doc=ProvJSONSerializer.deserialize(ProvJSONSerializer,doc)
    dot = prov_to_dot(doc)
    
    if format=="png":
        dot.write_png('PROV.png')
        return 'PROV.png' 
    if format=="svg":
        dot.write_svg('PROV.svg')
        return 'PROV.svg'
    
```

### Provenance of Creating Input Data Collection within a Data Volume (shared with one or more JupyterLab instances) via the execution of a workflow

```python
file = open("workflow_run.template.json", "r")
# Read the entire content of the file
content = file.read()

w = io.StringIO(content)
png_content=provJSONtoPNG(w,"png")


Image(png_content)
```
